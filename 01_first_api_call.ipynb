```python
from dotenv import load_dotenv
load_dotenv()

from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-6"  # latest model (course uses claude-sonnet-4-0)

print("Client ready!")
```

## Making your first request

`client.messages.create()` requires 3 key parameters:
- **model** — which Claude model to use
- **max_tokens** — a safety cap on response length (not a target)
- **messages** — the conversation history (list of dicts with `role` and `content`)

```python
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)

print(message.content[0].text)
```

## Inspecting the full response object

```python
print("Model:", message.model)
print("Stop reason:", message.stop_reason)
print("Input tokens:", message.usage.input_tokens)
print("Output tokens:", message.usage.output_tokens)
print("\nFull content block:", message.content)
```
